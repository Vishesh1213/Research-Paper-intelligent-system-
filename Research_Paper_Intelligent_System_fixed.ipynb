````python
import json
import time
import google.generativeai as genai
````

````python
GEMINI_API_KEY = "Your Key"

genai.configure(api_key=GEMINI_API_KEY)

model = genai.GenerativeModel("gemini-2.5-flash")
````

````python
keywords
````

````python
keyword_list = [k[0] for k in keywords]

print(keyword_list)
````

````python
def explain_keyword(keyword):

    prompt = f"""
You are an AI Research Assistant.

Explain the following technical keyword extracted from a research paper.

Keyword:
{keyword}

Return ONLY valid JSON.

{{
    "entity":"",
    "category":"",
    "definition":"",
    "importance":"",
    "applications":[],
    "related_terms":[]
}}
"""

    response = model.generate_content(prompt)

    try:
        text = response.text.strip()

        if text.startswith("```json"):
            text = text.replace("```json","").replace("```","").strip()

        return json.loads(text)

    except Exception:

        return {
            "entity": keyword,
            "category": "Unknown",
            "definition": response.text
        }
````

````python
entity_information = []

for word in keyword_list:

    print(f"Processing : {word}")

    try:

        info = explain_keyword(word)

        entity_information.append(info)

        time.sleep(2)

    except Exception as e:

        print(e)
````

````python
from pprint import pprint

for entity in entity_information:

    pprint(entity)

    print("-"*80)
````

````python
with open("paper_entities.json","w",encoding="utf-8") as f:

    json.dump(entity_information,
              f,
              indent=4,
              ensure_ascii=False)

print("Saved Successfully")
````
